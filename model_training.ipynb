```python

from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
from langdetect import detect
from googletrans import Translator
```

```python
model_name = "facebook/blenderbot-400M-distill"
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)
```

```python
translator = Translator()
```

```python
def chatbot_response(user_input):
    # Detect language
    lang = detect(user_input)

    # Translate to English if not already English
    if lang != "en":
        translated_input = translator.translate(user_input, src=lang, dest='en').text
    else:
        translated_input = user_input

    # Generate response in English
    inputs = tokenizer([translated_input], return_tensors='pt')
    reply_ids = model.generate(**inputs)
    reply_en = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

    # Translate back to original language
    if lang != "en":
        final_reply = translator.translate(reply_en, src='en', dest=lang).text
    else:
        final_reply = reply_en

    return final_reply, lang
```

```python
while True:
    user_msg = input("You: ")
    if user_msg.lower() in ["quit", "exit"]:
        break
    reply, lang_detected = chatbot_response(user_msg)
    print(f"[{lang_detected}] Bot:", reply)
```
